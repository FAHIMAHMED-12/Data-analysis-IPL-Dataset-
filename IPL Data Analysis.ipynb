```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
###Load Data
dataset=pd.read_csv("D:/Data Visualization Project/IPL Matches 2008-2020.csv")
```

```python
dataset.info()
```

```python
#Drop the method feature
dataset.drop(["method"],axis=1, inplace=True)
```

```python
dataset.info()
```

```python
#most wins in ipl(team)
wins = pd.DataFrame({"Winner": dataset["winner"]})
count_wins = wins.value_counts()
print(count_wins)
```

```python
#most wins in ipl in chart
wins = pd.DataFrame({"Winner": dataset["winner"]})
count_wins = wins.value_counts()
labels = [X[0] for X in count_wins.keys()]
bar,ax= plt.subplots(figsize=(20,12))
ax=plt.pie(x=count_wins, autopct = "%.1f%%" , labels=labels )
plt.title("Most wins in IPL", fontsize=17)
plt.show()
```

```python
#Most wins in Eliminator
sns.countplot(dataset["winner"][dataset["eliminator"]=="Y"], data=dataset)
plt.title("Most wins in Eliminator", fontsize=17)
plt.xticks(rotation=90)
plt.show()
```

```python
#toss decision
teams = dataset["toss_winner"].unique()
decision_making = pd.DataFrame([], columns= ["Toss Winner", "Decision", "Times"])
for id, element in enumerate(teams):
    team_bat = dataset[(dataset["toss_winner"]==element)&(dataset["toss_decision"]=="bat")]
    team_fields = dataset[(dataset["toss_winner"]==element)&(dataset["toss_decision"]=="field")]
#append to decision making
    decision_making=decision_making.append({"Toss Winner": element, "Decision": "bat", "Times": team_bat["toss_winner"].count()},ignore_index=True)
    decision_making=decision_making.append({"Toss Winner": element, "Decision": "fields", "Times": team_fields["toss_winner"].count()},ignore_index=True)
    
    
```

```python
decision_making
```

```python
sns.catplot(x="Toss Winner", y="Times", hue="Decision", data= decision_making, kind="bar", height=5, aspect=2)
plt.xticks(rotation=90)
plt.title("Toss decision of Teams")
plt.xlabel("IPL Teams")
plt.ylabel("Toss Decision")
```
